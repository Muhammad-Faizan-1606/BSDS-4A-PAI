```python
import cv2
import numpy as np
```

```python
img = cv2.imread('image.jpg')
print(img.shape)
print(type(img))
```

```python
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
h = img.shape[0]
w = img.shape[1]
print('height')
print(h)
print('width')
print(w)
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
print(gray.shape)
cv2.imshow('gray',gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imwrite('gray.jpg',gray)
print('saved')
```

```python
resized = cv2.resize(img,(300,300))
print(resized.shape)
cv2.imshow('resized',resized)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cropped = img[100:400,200:500]
print(cropped.shape)
cv2.imshow('cropped',cropped)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
rotated = cv2.rotate(img,cv2.ROTATE_90_CLOCKWISE)
print(rotated.shape)
cv2.imshow('rotated',rotated)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img2 = img.copy()
cv2.rectangle(img2,(50,50),(200,200),(0,255,0),3)
cv2.imshow('rectangle',img2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img3 = img.copy()
cv2.circle(img3,(100,100),50,(255,0,0),2)
cv2.imshow('circle',img3)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img4 = img.copy()
cv2.line(img4,(0,0),(300,300),(0,0,255),5)
cv2.imshow('line',img4)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img5 = img.copy()
cv2.putText(img5,'Hello',(50,50),cv2.FONT_HERSHEY_SIMPLEX,1,(255,255,255),2)
cv2.imshow('text',img5)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
blurred = cv2.GaussianBlur(img,(15,15),0)
cv2.imshow('blur',blurred)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
edges = cv2.Canny(img,100,200)
cv2.imshow('edges',edges)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
ret,thresh = cv2.threshold(gray,127,255,cv2.THRESH_BINARY)
print(ret)
cv2.imshow('threshold',thresh)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
kernel = np.ones((5,5),np.uint8)
eroded = cv2.erode(img,kernel,iterations=1)
cv2.imshow('eroded',eroded)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
dilated = cv2.dilate(img,kernel,iterations=1)
cv2.imshow('dilated',dilated)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
print(hsv.shape)
cv2.imshow('hsv',hsv)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
lower = np.array([0,50,50])
upper = np.array([10,255,255])
mask = cv2.inRange(hsv,lower,upper)
cv2.imshow('mask',mask)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
result = cv2.bitwise_and(img,img,mask=mask)
cv2.imshow('result',result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
contours,hierarchy = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
print('number of contours')
print(len(contours))
```

```python
img6 = img.copy()
cv2.drawContours(img6,contours,-1,(0,255,0),2)
cv2.imshow('contours',img6)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
faces = face_cascade.detectMultiScale(gray,1.1,4)
print('faces found')
print(len(faces))
```

```python
img7 = img.copy()
for i in range(len(faces)):
    x = faces[i][0]
    y = faces[i][1]
    w = faces[i][2]
    h = faces[i][3]
    cv2.rectangle(img7,(x,y),(x+w,y+h),(255,0,0),2)
cv2.imshow('faces',img7)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)
while True:
    ret,frame = cap.read()
    if ret == False:
        break
    cv2.imshow('video',frame)
    if cv2.waitKey(1) == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture('video.mp4')
while cap.isOpened():
    ret,frame = cap.read()
    if ret == True:
        cv2.imshow('frame',frame)
        if cv2.waitKey(25) == ord('q'):
            break
    else:
        break
cap.release()
cv2.destroyAllWindows()
```

```python
img8 = cv2.imread('image.jpg')
h = img8.shape[0]
w = img8.shape[1]
flipped = np.zeros((h,w,3),np.uint8)
for i in range(h):
    for j in range(w):
        flipped[i][j] = img8[i][w-1-j]
cv2.imshow('flipped',flipped)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img9 = cv2.imread('image.jpg')
h = img9.shape[0]
w = img9.shape[1]
negative = np.zeros((h,w,3),np.uint8)
for i in range(h):
    for j in range(w):
        for k in range(3):
            negative[i][j][k] = 255 - img9[i][j][k]
cv2.imshow('negative',negative)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img10 = cv2.imread('image.jpg')
h = img10.shape[0]
w = img10.shape[1]
total_b = 0
total_g = 0
total_r = 0
for i in range(h):
    for j in range(w):
        total_b = total_b + img10[i][j][0]
        total_g = total_g + img10[i][j][1]
        total_r = total_r + img10[i][j][2]
avg_b = total_b / (h*w)
avg_g = total_g / (h*w)
avg_r = total_r / (h*w)
print('average b')
print(avg_b)
print('average g')
print(avg_g)
print('average r')
print(avg_r)
```
